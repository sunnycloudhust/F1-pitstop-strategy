```python
import os

import pandas as pd
import fastf1

path = "/content/drive/Othercomputers/My Computer/Nguyen Tri/Code/Statisanalyss/f1_data_csv_export/2021/Dutch Grand Prix_Laps.csv"

df = pd.read_csv(path)
print(df.head())
print(df.columns)
print(df["Compound"].unique())

def drop_na_wet(df):
    # Drop rows where 'Compound' is NaN
    df = df.dropna(subset=['Compound'])
    # Filter out rows where 'Compound' is 'Wet' or 'Intermediate'
    df = df[~df['Compound'].isin(['Wet', 'Intermediate', 'UNKNOWN'])]
    return df

root = "/content/drive/Othercomputers/My Computer/Nguyen Tri/Code/Statisanalyss/f1_data_csv_export"
processed_root = "/content/drive/Othercomputers/My Computer/Nguyen Tri/Code/Statisanalyss/Preprocessed"
for folder in os.listdir(root):
    folder_path = os.path.join(root, folder)
    if os.path.isdir(folder_path):
        # Create the target directory if it doesn't exist
        os.makedirs(os.path.join(processed_root, folder), exist_ok=True)
        for file in os.listdir(folder_path):
            if file.endswith("_Laps.csv"):
                print(f"Processing: {file}")
                df = pd.read_csv(os.path.join(folder_path, file))
                df = drop_na_wet(df)
                df.to_csv(os.path.join(processed_root, folder, f"{os.path.splitext(file)[0]}_Laps_processed.csv"), index=False)
                print(df.head())
                print(df.columns)
                print(df["Compound"].unique())
```
